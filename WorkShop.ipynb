**Theorem** (Euler's summation formula). _If $ f $
has a continuous derivative $ f' $ on the interval
$ [y, x], $ where $ 0 < y < x, $ then_
\begin{align}
  \sum_{y < n \le x} f(n)
    = & \int_y^x f(t) \, dt + \int_y^x (t - [t]) f'(t) \, dt
       \notag \\
       & + f(x)([x] - x) - f(y)([y] - y). \label{theorem}
\end{align}

_Proof._ Let $ m = [y] $ and $ k = [x]. $ For
integers $ n $ and $ n - 1 $ in $ [y, x] $ we have
\begin{align*}
  \int_{n - 1}^n [t] f'(t) \, dt
    & = \int_{n - 1}^n (n - 1) f'(t) \, dt \\
    & = (n - 1) \{ f(n) - f(n - 1) \} \\
    & = \{n f(n) - (n - 1) f(n - 1)\} - f(n).
\end{align*}
Summing from $ n = m + 2 $ to $ n = k $ we find
that the first sum telescopes, hence
\begin{align*}
  \int_{m + 1}^k [t] f'(t) \, dt
    & = k f(k) - (m + 1) f(m + 1) - \sum_{n = m + 2}^k f(n) \\
    & = k f(k) - m f(m + 1) - \sum_{y < n \le x} f(n).
\end{align*}
Therefore
\begin{align}
  \sum_{y < n \le x} f(n)
    & = - \int_{m + 1}^k [t] f'(t) \, dt + k f(k) - m f(m + 1) \notag \\
    & = - \int_y^x [t] f'(t) \, dt + k f(x) - m f(y). \label{summation}
\end{align}
Integration by parts gives us
$$
\int_y^x f(t) \, dt = x f(x) - y f(y) - \int_y^x t f'(t) \, dt,
$$
